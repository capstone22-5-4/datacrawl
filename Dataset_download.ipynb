동물 목록 불러오기

```python
import pandas as pd
animal_list_csv = pd.read_csv('animals.csv')
animal_list = animal_list_csv.columns.tolist()
animal_list
```

```python
import os
import shutil
from bing_image_downloader.bing_image_downloader import downloader
from google_images_download.google_images_download import google_images_download as gim
```

동물 목록으로부터 구글에서 크롤링

```python
response = gim.googleimagesdownload()

arguments = {"keywords" : ','.join(animal_list), "limit":60, "print_urls":True}
paths = response.download(arguments)
```

폴더 생성

```python
folder_list = [
    './dataset/train/',
    './dataset/test/'
]

for folder in folder_list:
    if not os.path.isdir(folder):
        os.makedirs(folder)
        
def split_data(name, train_size):
    for folder in folder_list:
        if not os.path.isdir(folder+'/'+name):
            os.makedirs(folder + '/' + name)
            
    count= 0 
    for file_name in os.listdir('./downloads/'+name):
        if count < train_size:
            shutil.copy('./downloads/' + name+ '/' + file_name, './dataset/train/' + name + '/' + file_name)
        else:
            shutil.copy('./downloads/' + name+ '/' + file_name, './dataset/test/' + name + '/' + file_name)
        count += 1
#     shutil.rmtree(name)
```

## 40개는 train_data, 20 개는 test_data

```python
for animal in animal_list:
    split_data(animal, 40)
```
